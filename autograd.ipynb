```python
import torch
```

```python
x = torch.ones(5)
y = torch.zeros(3)
w = torch.randn(5, 3 ,requires_grad=True)
b = torch.randn(3, requires_grad=True)
```

```python
z = (x @ w) + b
loss = torch.nn.functional.binary_cross_entropy_with_logits(z, y)
```

```python
print(f"Gradient function for z = {z.grad_fn}")
print(f"Gradient function for loss = {loss.grad_fn}")
```

```python
loss.backward()
print(f"weights: {w.grad}")
print(f"bias: {b.grad}")
```

```python
#5. gradient accumulation
x = torch.tensor(2.0, requires_grad=True)

y3 = x**2

y3.backward()

print(x.grad)

y3 = x**2

y3.backward()

print(x.grad)
```

```python
x = torch.tensor(2.0, requires_grad=True)

y3 = x**2

y3.backward()

print(x.grad)

x.grad.zero_()  #makes the x = [0.0]

y3 = x**2

y3.backward()

print(x.grad)
```

```python
#7. multistep computational graph

x = torch.tensor(2.0, requires_grad=True)

a = x + 1
b = a**2
c = 2*b
d = c + 3
```

`x -> a = x+1 -> b = (x+1)**2 -> c = 2*((x+1)**2) -> d = 2*((x+1)**2)+3` 

```python
d.backward()
x.grad
```
