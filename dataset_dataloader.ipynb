```python
import torch
```

```python
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(
    n_samples=20,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    n_classes=2,
    random_state=42
)
```

```python
X.shape
```

```python
y.shape
```

```python
X_tensor = torch.tensor(X, dtype=torch.float32)
y_tensor = torch.tensor(y, dtype=torch.long)
```

```python
X_tensor
```

```python
y_tensor
```

```python
from torch.utils.data import Dataset, DataLoader
```

```python
class CustomDataset(Dataset):

    def __init__(self, features, labels):
        self.features = features
        self.labels = labels

    def __len__(self):
        return self.features.shape[0]

    def __getitem__(self, index):
        return self.features[index], self.labels[index]
```

```python
dataset = CustomDataset(X_tensor, y_tensor)
```

```python
len(dataset)   # or dataset.__len__()
```

```python
dataset[0]
```

```python
dataloader = DataLoader(dataset, batch_size=4, shuffle=True)
```

```python
for batch_features, batch_labels in dataloader:
    print(batch_features, '\n',batch_labels)
    print("-"*50)
```

```python
class LinearModel(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.layer_1 = torch.nn.Linear(2, 1)
        self.relu = torch.relu(self.layer_1())

    def forward(self, x):
        out = self.relu(self.layer_1(x))

        return out
```

```python
model_linear = LinearModel()
model_linear.forward(X_tensor)
```
