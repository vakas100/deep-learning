```python
def f(x):
    return x**2 + 2*x + 5
```

```python
f(2.0)
```

```python
xs = np.arange(-5, 5, 0.25)
ys = f(xs)
plt.plot(xs, ys, color='cyan', linestyle='-')
plt.title('Graph of function x^2+2x+5')
plt.xlabel('x values')
plt.ylabel('y values')
```

```python
class Value:
    def __init__(self, data):
        self.data = data

    def __repr__(self):
        return f"Value(data={self.data})"
    
    def __add__(self, other):
        return Value(self.data + other.data)
    
    def __mul__(self, other):
        return Value(self.data * other.data)
```

```python
a = Value(2.0)
b = Value(-3.0)
c = Value(10.0)
d = a * b + c
d
```
