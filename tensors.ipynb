## Random Tensors

### Random tensors are important because the way many neural networks work is they start with random numbers and then update them to better represent the data

`Start with random numbers -> look at data -> update the number -> look at data -> update the number`

```python
# create a random tensor of size of (3, 4)
RANDOM_TENSOR = torch.rand(3, 3, 4)
RANDOM_TENSOR
```

```python
RANDOM_TENSOR.ndim
```

```python
RANDOM_TENSOR.shape
```

```python
# creating a random tensor with the same shape as a image tensor
random_img_size_tensor = torch.rand(3, 224, 224)
random_img_size_tensor.shape, random_img_size_tensor.ndim
```
